```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("D:\\IGT\\data\\Halloween data set\\dataset\\train.csv")
```

```python
train_data.head()
```

```python
train_data.isnull().sum()
```

```python
train_data.drop(['instock_date','Customer_name'],axis =1, inplace = True)
```

```python
train_data.isnull().sum()
```

```python
train_data.shape
```

```python
train_data.drop(["Stall_no"],axis = 1, inplace = True)
```

```python
train_data.Discount_avail.unique()
```

```python
train_data['Discount_avail'].fillna(0, inplace = True)
```

```python
train_data.isnull().sum()
```

```python
train_data.describe()
```

```python
train_data.describe()
```

```python
train_data['charges_1'] = train_data['charges_1'].fillna(train_data['charges_1'].mean())
train_data['charges_2 (%)']= train_data['charges_2 (%)'].fillna(train_data['charges_2 (%)'].mean())
train_data['Minimum_price'] = train_data['Minimum_price'].fillna(train_data['Minimum_price'].mean())
train_data['Maximum_price'] = train_data['Maximum_price'].fillna(train_data['Maximum_price'].mean())
train_data['Selling_Price']= train_data['Selling_Price'].fillna(train_data['Selling_Price'].mean())
```

```python
train_data.isnull().sum()
```

```python
train_data.head()
```

```python
train_data['Loyalty_customer'] = train_data['Loyalty_customer'].apply(lambda x: x.strip())
```

```python
train_data.Product_id.unique()
```

```python
train_data.Product_Category.unique()
```

```python
train_data.Loyalty_customer.replace(to_replace = {'Yes' : 1, 'No' : 0}, inplace = True)
```

```python
train_data.Loyalty_customer.unique()
```

```python
final_train = pd.get_dummies(data=train_data, columns=['Product_Category', 'Product_id'], drop_first= True)
```

```python
final_train[final_train < 0] = 0
```

```python
final_train.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_train.drop(['Selling_Price'], axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
y = final_train.Selling_Price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
predicted
```

```python
for i in range(len(predicted)):
    if (predicted[i] < 0):
        print(predicted[i], end = "  ")
```

```python
y_test.shape
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_log_error
```

```python
RMLE = mean_squared_log_error(y_test, predicted)
```

```python
score = max(0, 100 - RMLE )
```
